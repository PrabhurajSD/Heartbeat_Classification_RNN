```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

import tensorflow as tf
```

**Data Loading**

```python
data = [pd.read_csv('../input/heartbeat/ptbdb_' + x + '.csv') for x in ['normal', 'abnormal']]
```

```python
data[0] #Normal dataframes
```

```python
data[1]#abnormal dataframes
```

**Data Preprocessing**

features [Columns] are not properly define, lets define them first

```python
for df in data:
    df.columns = list(range(len(df.columns)))
```

```python
data[1]
```

```python
fin_data = pd.concat(data, axis=0).sample(frac=1.0, random_state=1).reset_index(drop=True)
fin_data = fin_data.rename({187: 'Label'}, axis=1)
```

```python
fin_data
```

```python
fin_data.describe()
```

```python
fin_data.info()
```

```python
y = fin_data['Label'].copy()
X = fin_data.drop('Label', axis=1).copy()
```

```python
y
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=1)
```

```python
X_train
```

```python
y_train
```

**Model Building**

```python
inputs = tf.keras.Input(shape=(X_train.shape[1],))

expand = tf.expand_dims(inputs, axis=2)
gru = tf.keras.layers.GRU(256, return_sequences=True)(expand)
flatten = tf.keras.layers.Flatten()(gru)

outputs = tf.keras.layers.Dense(1, activation='sigmoid')(flatten)


model = tf.keras.Model(inputs=inputs, outputs=outputs)

print(model.summary())
```

**Training**

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc')
    ]
)

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=32,
    epochs=10,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
    ]
)
```

**Testing/Results**

```python
results = model.evaluate(X_test, y_test, verbose=0)
```

```python
print("Test Accuracy: {:.2f}%".format(results[1] * 100))
print("     Test AUC: {:.4f}".format(results[2]))
```
